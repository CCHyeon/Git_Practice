## 리스트 항목 삭제 (del, remove, pop)

```python
del_list=['a', 'b', 'c', 'd', 'f', 'e']

del del_list[-1]

print(del_list)
```

```python
del_list=['a', 'b', 'c', 'd', 'f', 'e', 'f', 'f', 'f']

del_list.remove(del_list[-1])
del_list.remove('f')

print(del_list)
```

```python
del_list=['a', 'b', 'c', 'd', 'f', 'e']

del_list.pop(0)

print(del_list)
```
