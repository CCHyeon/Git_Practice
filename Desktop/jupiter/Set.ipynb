### 집합의 연산

```python
union_set = {1,2,3} | {2,3,4}
print(union_set)
```

```python
union_set = {1,2,3}.union({2,3,4})
print(union_set)
```

```python
diff_set = {1,2,3} - {2,3,4}
print(diff_set)
```

```python
diff_set = {1,2,3}.diff({2,3,4})
print(diff_set)
```

```python
intersec_set = {1,2,3} & {2,3,4}
print(intersec_set)
```

```python
#대칭차집합
sys_set = {1,2,3} ^ {2,3,4}
print(sys_set)
```
